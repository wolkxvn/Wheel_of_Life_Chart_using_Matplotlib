```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.text as txt

plt.figure(figsize=(10,10), facecolor='#0C0014')

ax = plt.gca()
spines = ax.spines
for spine in spines.values():
    spine.set_visible(False)
ax.set_aspect('equal')
ax.set_facecolor('#0C0014')

ax.xaxis.set_major_locator(plt.NullLocator())  # Hide x-axis major ticks
ax.yaxis.set_major_locator(plt.NullLocator())  # Hide y-axis major ticks

# set radius and theta
theta = np.linspace(0, 2*np.pi, 8)

n = 10

classes = {'Career':6, 'Financial':4, 'Spiritual':8, 'Physical':7, 'Intellectual':3, 'Family':10, 'Social':8}

hex_colors = [
    "#9C27B0"  ,  # Violet
    "#4CAF50"  ,  # Green
    "#2196F3"  ,  # Light Blue
    "#F44336"  ,  # Red
    "#FFEB3B"  ,  # Yellow
    "#E91E63"  ,  # Hot Pink
    "#00BCD4"    # Teal
]

lx = []
ly = []

for cls, cls_value, te, hex_color in zip(classes.keys(),classes.values(), np.linspace(0, 2*np.pi, (len(classes)+1)), hex_colors):

  x1, y1 = n * np.cos(te), n * np.sin(te)

  x_text, y_text = (n + 1.5) * np.cos(te), (n + 1.5) * np.sin(te)

  x_v, y_v = cls_value * np.cos(te), cls_value * np.sin(te)
  lx.append(x_v)
  ly.append(y_v)

  ax.plot([0,x1], [0, y1], color = '#868686')

  ax.text(x=x_text, y=y_text, s=cls, fontsize=13, fontweight='bold', ha='center', va='center', color=hex_color)

  ax.plot(x_v, y_v, marker='o', color=hex_color)
  ax.plot(x1, y1, marker='o', color=hex_color)


print(lx, ly)

ax.fill(lx, ly, color='#FF9510', alpha=0.15)
lx.append(lx[0])
ly.append(ly[0])
ax.plot(lx, ly, color='#FF9510')


for radius in range(n):
  a = radius * np.cos( theta )
  b = radius * np.sin( theta )

  if radius %2 == 0:
    ax.plot(a, b, color = '#868686', alpha=0.3)
    continue
  # plot in red color
  ax.plot(a, b, color = '#868686')

# set limits slightly bigger than radius
plt.xlim(-1.5 * n, 1.5 * n)
plt.ylim(-1.5 * n, 1.5 * n)

plt.title("Wheel of Life", fontsize=16, fontweight='bold', ha='center', va='center', color='#6F6AC4')
plt.savefig("Wheel-of-life.png", dpi=300, format='png')
plt.show()
```
